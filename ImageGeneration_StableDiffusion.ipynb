```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
```

### **Using pre-trained Stable Diffusion Model**

```python
model_id ="stabilityai/stable-diffusion-2"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16, use_safetensors=True).to("cuda")
```

### ***Prompt-1***

```python
prompt = """a girl,beautiful, travelling between dimensions, dynamic pose, soft eyes and narrow chin,
 dainty figure, short hair straight down, torn kawaii shirt and baggy jeans
"""
```

```python
image = pipe(prompt, width=400, height=800).images[0]
```

```python
print("Here is the picture on the basis of the given prompt")
plt.imshow(image);
plt.axis('off');
```

### **Prompt-2**

```python
prompt = "dreamlike, school boy, showing excitement, wearing school uniform, holding his water bottle in his hands"
```

```python
image = pipe(prompt, width=800, height=800,num_images_per_prompt=2).images
```

```python
print("Here is the picture on the basis of the given prompt")

for i in image:
    plt.imshow(i)
    plt.axis('off')
    plt.show()
```
